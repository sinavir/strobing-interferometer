## 2 Tuning the analysis

```python
import numpy as np

with h5py.File(STROBE_IMG_PATH + FILE, 'r') as f:
    image = a.StdAnalysis(f)

    # Tune the smoothing kernel for the calibration smoothing
    image.smooth_calibration(window=np.array([0.1, 0.25, 0.3, 0.25, 0.1]))

    image.compute_calibration_slopes()
    image.compute_independant_video_images()
    image.combine_images()

    # Tuning the masking parameters
    #   See the reference for more explanations (https://sinavir.github.io/strobing-interferometer/)
    image.apply_membrane_shape_masking(threshold = 1.0, sigma = 100.0)

    # percentile: the percentile above which we clip the values (ex: 99% means the 1% extreme values are clipped)
    image.clip_high_values(percentile = 99.0)
```
